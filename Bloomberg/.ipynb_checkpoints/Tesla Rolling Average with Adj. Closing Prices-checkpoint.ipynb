```python
import quandl
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
plt.rcParams['figure.figsize'] = [15, 15]
```

```python
tsla = quandl.get("WIKI/TSLA", start_date = "2003-01-01", end_date = "2020-01-01")
```

```python
tsla['100ma'] = tsla['Adj. Close'].rolling(window = 100, min_periods = 0).mean()
```

```python
tsla
```

```python
ax1 = plt.subplot2grid((6,1), (0,0), rowspan = 5, colspan = 1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan = 5, colspan = 1, sharex=ax1)

ax1.plot(tsla.index, tsla['Adj. Close'])
ax1.plot(tsla.index, tsla['100ma'])
ax2.bar(tsla.index, tsla['Volume'])

plt.show()
```
